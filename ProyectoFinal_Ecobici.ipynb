# 2012.

## Total de Usuarios. Segmentado por Género.

```python
#Usamos el dataframe rangos_horas_generos

df_rangos_horas_generos_2012 = pd.read_csv("df_rangos_horas_generos_2012.csv")

fig = go.Figure()
fig.add_trace(go.Bar(
    x=df_rangos_horas_generos_2012["Horas"],
    y=df_rangos_horas_generos_2012["Usos_Gen_F"],
    name='Mujeres',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=df_rangos_horas_generos_2012["Horas"],
    y=df_rangos_horas_generos_2012["Usos_Gen_M"],
    name='Hombres',
    marker_color='lightsalmon'
))

# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_layout(barmode='group',
                  xaxis_tickangle=-45,
                  title="Usuarios por género. Año 2012.",
                  xaxis_title="Horas",
                  yaxis_title="Total de usuarios.")
fig.show()
```

## Rangos de tiempo de uso de bici por día.

```python
#Usamos el dataframe rangos_tiempo_uso_dias

df_rangos_tiempo_uso_dias_2012 = pd.read_csv("df_rangos_tiempo_uso_dias_2012.csv")

fig = go.Figure()
fig.add_trace(go.Bar(
    x=df_rangos_tiempo_uso_dias_2012["Rangos_T.Uso"],
    y=df_rangos_tiempo_uso_dias_2012["Usos_Lunes"],
    name='Uso_Lunes',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=df_rangos_tiempo_uso_dias_2012["Rangos_T.Uso"],
    y=df_rangos_tiempo_uso_dias_2012["Usos_Martes"],
    name='Uso_Martes',
    marker_color='lightsalmon'
))
fig.add_trace(go.Bar(
    x=df_rangos_tiempo_uso_dias_2012["Rangos_T.Uso"],
    y=df_rangos_tiempo_uso_dias_2012["Usos_Miércoles"],
    name='Uso_Miércoles',
    #marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=df_rangos_tiempo_uso_dias_2012["Rangos_T.Uso"],
    y=df_rangos_tiempo_uso_dias_2012["Usos_Jueves"],
    name='Uso_Jueves',
    #marker_color='lightsalmon'
))
fig.add_trace(go.Bar(
    x=df_rangos_tiempo_uso_dias_2012["Rangos_T.Uso"],
    y=df_rangos_tiempo_uso_dias_2012["Usos_Viernes"],
    name='Uso_Viernes',
    #marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=df_rangos_tiempo_uso_dias_2012["Rangos_T.Uso"],
    y=df_rangos_tiempo_uso_dias_2012["Usos_Sábado"],
    name='Uso_Sábado',
    #marker_color='lightsalmon'
))
fig.add_trace(go.Bar(
    x=df_rangos_tiempo_uso_dias_2012["Rangos_T.Uso"],
    y=df_rangos_tiempo_uso_dias_2012["Usos_Domingo"],
    name='Uso_Domingo',
    #marker_color='lightsalmon'
))
# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_layout(barmode='group',
                  xaxis_tickangle=-45,
                  title="Usuarios por rangos de tiempo de uso del servicio. Seccionado por día. Año 2012.",
                  xaxis_title="Tiempo de uso. (Menor a.. )",
                  yaxis_title="Total de usuarios.")
fig.show()
```

## Total de usos por rangos de edad y género.

```python
#Usamos el dataframe rangos_edades_generos

df_rangos_edades_generos_2012 = pd.read_csv("df_rangos_edades_generos_2012.csv")

fig = go.Figure()
fig.add_trace(go.Bar(
    x=df_rangos_edades_generos_2012["Edades (Menor a)"],
    y=df_rangos_edades_generos_2012["Usos_Gen_F"],
    name='Mujeres',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=df_rangos_edades_generos_2012["Edades (Menor a)"],
    y=df_rangos_edades_generos_2012["Usos_Gen_M"],
    name='Hombres',
    marker_color='lightsalmon'
))

# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_layout(barmode='group',
                  xaxis_tickangle=-45,
                  xaxis = dict(
                      tickmode = 'array',
                      tickvals = df_rangos_edades_generos_2012["Edades (Menor a)"],
                      ticktext = ['Entre 16 y 18', 'Entre 18 y 20', 'Entre 20 y 25', 'Entre 25 y 30', 'Entre 30 y 35', 'Entre 35 y 40', 'Entre 40 y 50', 'Entre 50 y 60', 'Mas de 60']
                      ),
                  title="Usuarios por rangos de edad divididos por género. Año 2012.",
                  xaxis_title="Edades (Años)",
                  yaxis_title="Total de usuarios.")
fig.show()
```
